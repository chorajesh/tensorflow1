### Classification Methods  in Python with Scikit-Learn

######  Module 1: Introduction to Classification; 

Machine Learning allows computers to learn and infer from data

In machine learning, classification refers to a predictive modeling problem where a class label is predicted for a given example of input data.

Types of Machine Learning

      Supervised
     
      Unsupervised
      
Supervised Learning : Data points have a known outcome

Unsupervised Learning : Data points have unknown outcome


Types of Supervised Learning: 

Regression        : Outcome is continuous (numerical)   
 
                    Ex:- home prices,  happiness index 

Classification    : Outcome is a Category
                    
                    Ex:- Object classes in Images

######  Module 2: Importing Libraries and Datasets

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
df.head(51)
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['target'].value_counts()
```

```python
df['target'].value_counts().plot.bar()
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']

X = df.loc[:, features].values
X
```

```python
y = df.loc[:,['target']].values
y
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(X)
#x_sc
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=30)
```

```python
x_train
```

```python
x_test
```

```python
import numpy as np
```

```python
y_train = np.ravel(y_train, order='C')
```

```python
y_test
```

```python
# import warnings
# warnings.filterwarnings('ignore')
```

######  Module 3: K-Nearest Neighbors;

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
np.ravel(y_train,order='C')
```

```python
knn = KNeighborsClassifier()
clf_knn= knn.fit(x_train,y_train)
```

```python
y_pred=clf_knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

print("\nAccuracy score: %f" %(accuracy_score(y_test,y_pred) * 100))
```

```python
from sklearn.metrics import precision_score, recall_score, classification_report,f1_score, confusion_matrix
print("Precision Score %f"%(precision_score(y_test,y_pred, average='micro')*100))
print("Recall Score %f"%(recall_score(y_test,y_pred, average='micro')*100))
print("F1 Score %f"%(f1_score(y_test,y_pred, average='micro')*100))
print(classification_report(y_test, y_pred, labels=['Iris-setosa','Iris-versicolor', 'Iris-virginica']))
print(confusion_matrix(y_test,y_pred,labels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']))
```

```python
probas = clf_knn.predict_proba(x_test)

import matplotlib.pyplot as plt
plt.figure(dpi=100)
plt.hist(probas, bins=10)
plt.title('Classification Probabilities')
plt.xlabel('Probability')
plt.ylabel('# of Instances')
plt.xlim([0.5, 1.0])
plt.legend(y_test)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred)) 
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
